# <p style="background-color:midnightblue; font-family:calibri; font-size:130%; color:white; text-align:center; border-radius:10px 10px; padding:15px"> Step 9: Concatenate arrays</p>

```python
a1=np.array([[1,2,3],
            [4,5,6]])

a2=np.array([[7,8,9],
            [10,11,12]])
np.hstack((a1, a2))
```

```python
a1=np.array([[1,2],[3,4],[5,6]])

a2=np.array([[7,8],[9,10]])

np.vstack((a1,a2))
```
